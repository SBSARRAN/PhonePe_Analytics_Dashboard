```python
# Extracting Map_Transaction from Phone_Pay_pluse
import os
path_Map_Trans = "C:/Phone_pay/pulse/data/map/transaction/hover/country/india/state/"
Agg_state_list=os.listdir(path_Map_Trans)
Agg_state_list
```

```python
import pandas as pd
import json
import os

clm={'State':[], 'Year':[],'Quater':[],'Map_Transacion_District_Name':[], 'Map_Transacion_count':[], 'Map_Transacion_amount':[]}

for i in Agg_state_list:
    p_i=path_Map_Trans+i+"/"
    Agg_yr=os.listdir(p_i)
    for j in Agg_yr:
        p_j=p_i+j+"/"
        Agg_yr_list=os.listdir(p_j)
        for k in Agg_yr_list:
            p_k=p_j+k
            Data=open(p_k,'r')
            D=json.load(Data)
            for z in D['data']['hoverDataList']:
              Name=z['name']
              count=z['metric'][0]['count']
              amount=z['metric'][0]['amount']
              clm['Map_Transacion_District_Name'].append(Name)
              clm['Map_Transacion_count'].append(count)
              clm['Map_Transacion_amount'].append(amount)
              clm['State'].append(i)
              clm['Year'].append(j)
              clm['Quater'].append(int(k.strip('.json')))
#Succesfully created a dataframe
Map_Trans=pd.DataFrame(clm)
```

```python
# Extracting Map_Users from Phone_Pay_pluse
import os
path_Map_Users = "C:/Phone_pay/pulse/data/map/user/hover/country/india/state/"
Agg_state_list=os.listdir(path_Map_Users)
Agg_state_list
```

```python
import pandas as pd
import json
import os

clm = {'State': [], 'Year': [], 'Quater': [], 'Map_District_Users': [], 'Map_Registered_Users': [],'Map_AppOpens_Users': []}

for i in Agg_state_list:   # Each state
    p_i = path_Map_Users + i + "/"
    Agg_yr = os.listdir(p_i)
    
    for j in Agg_yr:   # Each year
        p_j = p_i + j + "/"
        Agg_yr_list = os.listdir(p_j)
        
        for k in Agg_yr_list:   # Each quarter file
            p_k = p_j + k
            with open(p_k, 'r') as Data:
                D = json.load(Data)
                print(D)
            
            if "data" in D and "hoverData" in D["data"]:
                 for district, values in D["data"]["hoverData"].items():
                    clm['Map_District_Users'].append(district)
                    clm['Map_Registered_Users'].append(values.get('registeredUsers', 0))
                    clm['Map_AppOpens_Users'].append(values.get('appOpens', 0))
                    clm['State'].append(i)
                    clm['Year'].append(j)
                    clm['Quater'].append(int(k.strip('.json')))
#Succesfully created a dataframe
Map_Users=pd.DataFrame(clm)


```
